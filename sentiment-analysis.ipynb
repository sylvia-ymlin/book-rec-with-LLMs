```python
import numpy as np
emotion_labels = ["anger", "disgust", "fear", "joy", "neutral", "sadness", "surprise"]
isbn = [] # the unique book identifier
emotion_scores = {label: [] for label in emotion_labels}

# the prediction for a single book description, is the max score for each emotion across all sentences
def calculate_max_emotion_scores(predictions):
    per_emotion_scores = {label: [] for label in emotion_labels}
    for prediction in predictions:
        sorted_prediction = sorted(prediction, key=lambda x: x["label"])
        for index, label in enumerate(emotion_labels):
            per_emotion_scores[label].append(sorted_prediction[index]["score"])

    return {label: np.max(scores) for label, scores in per_emotion_scores.items()}
```

```python
for i in range(10):
    isbn.append(books["isbn13"][i])
    sentences = books["description"][i].split(".") # split the description into sentences
    predictions = classifier(sentences) # get predictions for each sentence
    max_emotion_scores = calculate_max_emotion_scores(predictions) # get the max emotion scores across all sentences
    for label in emotion_labels:
        emotion_scores[label].append(max_emotion_scores[label])
```

```python
from tqdm import tqdm

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = [] # the unique book identifier
emotion_scores = {label: [] for label in emotion_labels}

# score all book descriptions
for i in tqdm(range(len(books))):
    isbn.append(books["isbn13"][i])
    sentences = books["description"][i].split(".") # split the description into sentences
    predictions = classifier(sentences) # get predictions for each sentence
    max_emotion_scores = calculate_max_emotion_scores(predictions) # get the max emotion scores across all sentences
    for label in emotion_labels:
        emotion_scores[label].append(max_emotion_scores[label])
```

```python
emotions_df = pd.DataFrame(emotion_scores)
emotions_df["isbn13"] = isbn
```

```python
emotions_df
```

```python
# merge into the original books dataframe
books_ = pd.merge(books, emotions_df, on = "isbn13")
books_
```

```python
# change the columns name, the neutral column to sadness, and sadness to surprise and then surprise to neutral
books_ = books_.rename(columns={"neutral": "neutral_temp", "sadness": "sadness_temp", "surprise": "neutral"})
books_ = books_.rename(columns={"neutral_temp": "sadness"})
books_ = books_.rename(columns={"sadness_temp": "surprise"})
books_
```

```python
books_.to_csv("books_with_emotions.csv", index = False)
```
